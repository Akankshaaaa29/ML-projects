```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.math import confusion_matrix
```

```python
import cv2
```

```python
(X_train,y_train),(X_test,y_test)= mnist.load_data()
```

```python
type(X_train)
```

```python
X_train.shape
```

```python
print(X_train[10])
```

```python
plt.imshow(X_train[50])
plt.show()
print(y_train[50])
```

```python
np.unique(y_train)
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),keras.layers.Dense(50,activation='relu'),keras.layers.Dense(50,activation='relu'),keras.layers.Dense(10,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
loss,accuracy=model.evaluate(X_test,y_test)
```

```python
plt.imshow(X_test[0])
plt.show
```

```python
y_test[0]
```

```python
y_pred=model.predict(X_test)
```

```python
print(y_pred[0]) #gives probability
```

```python
label=np.argmax(y_pred[0]) #gives the label of the maximum number of y pred
```

```python
label
```

```python
y_pred_label=[np.argmax(i) for i in y_pred]
y_pred_label
```

```python
conf_mat=confusion_matrix(y_test,y_pred_label)
```

```python
conf_mat
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(conf_mat,annot=True,fmt='d',cmap='crest')
plt.ylabel("true labels")
plt.xlabel("predicted labels")
```
